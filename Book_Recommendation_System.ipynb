```python
AAS_RATINGS = pd.read_csv(f"{path}/Ratings.csv")
print("Ratings dataset loaded successfully.")
AAS_RATINGS.head(5)
```

```python
AAS_RATINGS.shape
```

```python
print(f"Shape of AAS_BOOKS: {AAS_BOOKS.shape}")
print(f"Shape of AAS_USERS: {AAS_USERS.shape}")
print(f"Shape of AAS_RATINGS: {AAS_RATINGS.shape}")
```

```python
AAS_RATINGS['User-ID'].value_counts()
```

```python
#filtering users who have rated less then 200 books
AAS_RATINGS_Valid= AAS_RATINGS['User-ID'].value_counts() > 200
```

```python
AAS_RATINGS_Valid[AAS_RATINGS_Valid].shape
```

```python
AAS_Valid_User_IDs = AAS_RATINGS_Valid[AAS_RATINGS_Valid].index
AAS_Filtered_Ratings = AAS_RATINGS[AAS_RATINGS['User-ID'].isin(AAS_Valid_User_IDs)]
AAS_Filtered_Ratings.head(5)
```

```python
AAS_Filtered_Ratings.shape
```

```python
AAS_Ratings_Books = AAS_Filtered_Ratings.merge(AAS_BOOKS, on='ISBN')
```

```python
AAS_Ratings_Books.head()
```

```python
AAS_Ratings_Books.shape
```

```python
AAS_num_rating = AAS_Ratings_Books.groupby('Book-Title')['Book-Rating'].count().reset_index()
```

```python
AAS_num_rating.head()
```

```python
AAS_num_rating.rename(columns={'Book-Rating': 'num_of_rating'}, inplace=True)
AAS_num_rating.head()
```

```python
AAS_final_ratings = AAS_Ratings_Books.merge(AAS_num_rating, on='Book-Title')
```

```python
AAS_final_ratings.head()
```

```python
AAS_final_ratings.shape
```

```python
# Taking Books with more then 50 Ratings
AAS_final_ratings = AAS_final_ratings[AAS_final_ratings['num_of_rating'] >= 50]
```

```python
AAS_final_ratings.head()
```

```python
AAS_final_ratings.shape
```

```python
AAS_final_ratings.drop_duplicates(['User-ID', 'Book-Title'], inplace=True)
```

```python
AAS_final_ratings.shape
```

```python
## Matrix Formation
AAS_matrix= AAS_final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
AAS_matrix
```

```python
AAS_matrix.shape
```

```python
AAS_matrix.fillna(0, inplace=True)
```

```python
AAS_matrix
```

```python
from scipy.sparse import  csr_matrix
```

```python
AAS_matrix_sparse = csr_matrix(AAS_matrix)
```

```python
AAS_matrix_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(AAS_matrix_sparse)
```

```python
distance, suggestion = model.kneighbors(AAS_matrix.iloc[237, :].values.reshape(1, -1), n_neighbors=10)
```

```python
import pickle
import os

# Create the 'artifacts' directory if it doesn't exist
if not os.path.exists('artifacts'):
    os.makedirs('artifacts')

pickle.dump(model, open('artifacts/model.pkl', 'wb'))
pickle.dump(AAS_book_name, open('artifacts/AAS_book_name.pkl', 'wb'))
pickle.dump(AAS_final_ratings, open('artifacts/AAS_final_ratings.pkl', 'wb'))
pickle.dump(AAS_matrix, open('artifacts/AAS_matrix.pkl', 'wb'))
```

```python
def recommend_book(AAS_book_name):
  book_id = np.where(AAS_matrix.index == AAS_book_name)[0][0]
  distance, suggestion = model.kneighbors(AAS_matrix.iloc[book_id, :].values.reshape(1, -1), n_neighbors=10)

  print(f"Recommendations for {AAS_book_name}:")
  for i in range(len(suggestion)):
    books = AAS_matrix.index[suggestion[i]]
    for j in books:
      print(j)
```
